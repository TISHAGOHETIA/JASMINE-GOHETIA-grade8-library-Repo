```python
import requests
import json
"""
The 'book' variable is already loaded from the URL in a previous cell.
Therefore, we can directly use the 'books' variable.
"""
data = book
print("JSON file loaded")
ref = db.reference("books")
for book in data["books"]:
  ref.child(str(book["id"])).set(book)
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
for books_id, info in books_data.items():
    print(info["title"])
```

```python
for books_id, info in books_data.items():
    print(info["title"], "-", info["details"]["available"])
```

```python
copies_list = []

for books_id, info in books_data.items():
    copies_list.append(info["details"]["copies"])

total_books = sum(copies_list)

print("Total number of books:", total_books)
```
